### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']  # Different colors for each bar

# Plot
plt.bar(categories, values, color=colors)

# Customizations
plt.title("Bar Plot of Categories")      # Title
plt.xlabel("Categories")                 # X-axis label (optional)
plt.ylabel("Values")                     # Y-axis label (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Horizontal grid lines for better readability

# Display the plot
plt.show()
```
